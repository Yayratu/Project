```python
import pandas as pd

def year_fraction(start_date, end_date, day_convention):
    start_date = pd.to_datetime(start_date, format='%d.%m.%Y')
    end_date = pd.to_datetime(end_date, format='%d.%m.%Y')
    
    if day_convention.lower() == 'act/365f':
        days_in_year = 365.0
    elif day_convention.lower() == 'act/360':
        days_in_year = 360.0
    elif day_convention.lower() == '30/360':
         days_in_year = 360.0
    elif day_convention.lower() == '30e/360':
         days_in_year = 360.0
    elif day_convention.lower() == 'act/act':
         days_in_year = (end_date - start_date).days
    elif day_convention.lower() == 'icma':
         days_in_year = 365.25
    elif day_convention.lower() == 'bd/252':
         days_in_year = 252.0
    else:
        raise ValueError('Unsupported day count convention')
    
    return (end_date - start_date).days / days_in_year

# Example dates
start_date = '15.2.2012'
end_date = '31.12.2012'
conventions = ['Act/365f', 'Act/360', '30/360', '30E/360', 'Act/Act', 'ICMA', 'BD/252']

# Calculate days between start and end date
days_difference = (pd.to_datetime(end_date, format='%d.%m.%Y') - pd.to_datetime(start_date, format='%d.%m.%Y')).days

# Generating data using year_fraction function for each convention
data = {'Day Count Convention': conventions,
        'Days': [days_difference] * len(conventions),  # Repeat the days_difference for each convention
        'Year': [year_fraction(start_date, end_date, conv) for conv in conventions]}

# Creating DataFrame from the generated data
df = pd.DataFrame(data).set_index('Day Count Convention')
 df = pd.DataFrame(data)
return df.rename_axis(None).reset_index(drop=True)
```
